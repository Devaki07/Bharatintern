```python
#importing necessary libraries
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#load dataset
```

```python
house_predict=pd.read_csv(r"C:\Users\pujit\Desktop\509\kc.csv")
house_predict
```

```python
#splitting dataset into dependent(x)variables and inpendent variables(y)
```

```python
features=['sqft_living','bedrooms']
```

```python
x=house_predict[features]
y=house_predict['price']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
plt.scatter(x['sqft_living'],y)
plt.xlabel('Square foot')
plt.ylabel('Price')
plt.title('Relationship between square foot and Price')
plt.show()
```

```python
#training the model
```

```python
x_train=pd.DataFrame(x_train,columns=features)
x_test=pd.DataFrame(x_test,columns=features)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#predicting the values
```

```python
y_pred=model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
print(f"mean squared error : {mse} ")
```

```python
new=[[2000,4]]
predicted_price=model.predict(new)
print(f"Predicted Price for New House :{predicted_price}")
```
